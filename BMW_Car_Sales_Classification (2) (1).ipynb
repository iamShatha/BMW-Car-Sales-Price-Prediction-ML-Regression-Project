```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('BMW_Car_Sales_Classification.csv')
```

```python
#data=data.drop('Sales_Classification',axis=1)
```

```python
data.info()
```

```python
data= pd.get_dummies(data,columns=['Sales_Classification','Transmission','Fuel_Type','Color','Region','Model'])
```

```python
data.describe()
```

```python
x=data.drop('Price_USD',axis=1)
y=data['Price_USD']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=23)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
c=pd.DataFrame(lm.coef_,x.columns,columns=['Price_USD'])
```

```python
c
```

```python
pr=lm.predict(x_test)
```

```python
pr
```

```python
y_test
```

```python
plt.scatter(x=pr,y=y_test)
```

```python
from sklearn import metrics 
```

```python
metrics.mean_absolute_error(y_test,pr)
```

```python
metrics.mean_squared_error(y_test,pr)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,pr))
```

```python
metrics.r2_score(y_test,pr)
```

```python
sns.displot(y_test-pr,bins=25)
```
